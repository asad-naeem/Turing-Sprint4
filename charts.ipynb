```python
fig = plt.figure()

# Add subplots to the figure
ax1 = fig.add_subplot(221)  # 1st subplot in a 2x2 grid
ax2 = fig.add_subplot(222)  # 2nd subplot in a 2x2 grid
ax3 = fig.add_subplot(223)  # 3rd subplot in a 2x2 grid
ax4 = fig.add_subplot(224)  # 4th subplot in a 2x2 grid

# Plot data in each subplot
# ax1.plot([1, 2, 3], [4, 5, 6])
# ax2.plot([1, 2, 3], [6, 5, 4])
# ax3.plot([1, 2, 3], [4, 6, 5])
# ax4.plot([1, 2, 3], [5, 4, 6])
ax1.hist(np.random.standard_normal(100), bins=20, color="black", alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.standard_normal(30))
ax3.plot(np.random.standard_normal(50).cumsum(), color="black",linestyle="dashed")
ax4.plot(np.random.standard_normal(30).cumsum(), color="black",linestyle="dashed", marker="o")

# Show the figure with multiple subplots
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.standard_normal(500), bins=50,
                        color="black", alpha=0.5)
fig.subplots_adjust(wspace=0, hspace=0)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
data = np.random.standard_normal(30).cumsum()
ax.plot(data, color="black", linestyle="dashed", label="Default")
ax.plot(data, color="black", linestyle="dashed",drawstyle="steps-post", label="steps-post")
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [4, 5, 6], marker='o')

# Annotate the second point
ax.annotate('Second point', xy=(2, 5))

plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [4, 5, 6], marker='o')

# Annotate the second point with an arrow
ax.annotate('Second point', xy=(2, 5), xytext=(3, 4),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [4, 5, 6], marker='o')

# Annotate the second point with customized text and arrow properties
ax.annotate('Second point', xy=(2, 5), xytext=(3, 4),
            arrowprops=dict(facecolor='black', arrowstyle='->'),
            fontsize=12, color='red', horizontalalignment='right', verticalalignment='top')

plt.show()
```

```python
fig, ax = plt.subplots()

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color="black", alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color="blue", alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color="green", alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```
